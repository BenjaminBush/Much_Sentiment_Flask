```python
i_urls = []
b_urls = []
p_urls = []
for link in urls:
    soup = BeautifulSoup(urllib2.urlopen(link), 'lxml')
    i_check = False
    b_check = False
    
    for val in soup.find('span', class_='displaytext').find_all('p'):
        if val.find('i') is not None:
            i_check = True
            break
        if val.find('b') is not None:
            b_check = True
            break

    if i_check:
        i_urls.append(link)
    elif b_check:
        b_urls.append(link)
    else:
        p_urls.append(link)
i_urls.remove("http://www.presidency.ucsb.edu/ws/index.php?pid=29407")
p_urls.append("http://www.presidency.ucsb.edu/ws/index.php?pid=29407")
```

```python
for link in p_urls:
    soup = BeautifulSoup(urllib2.urlopen(link), 'lxml')
    people_array = dates_namearray_dict[url_dates_dict[link]]
    content_dict = {}
    other_matches = {} #take this out after
    name_found = 0
    for name in people_array:
        content_dict[name] = ""
    recent_speaker = people_array[0]
    for val in soup.find('span', class_='displaytext').find_all('p'):
        my_text = val.get_text()
        real_p_text = ''
        real_string = ''
        name_found = 0
        if len(my_text.split()) > 5:
            mrcount = 0
            new_separator = my_text.split()[:3]
            for x in new_separator:
                if ":" in x:
                    real_string, real_p_text = my_text.split(":")[0:2]
                    break
                elif "." in x:
                    if mrcount:
                        if x[:-1].title() != "Mr" and x[:-1].title() != "Mrs" and x[:-1].title() != "Ms" and x[:-1].title() != "no" and x[:-1].title() != "Yes" and x[:-1].title() != "Sorry":
                            real_string = (' '.join(new_separator[:new_separator.index(x)+1]))[:-1]
                            real_p_text = ' '.join((my_text.split(x))[1:])
                            break
                    else:
                        if x[:-1].title() == "Mr" or x[:-1].title() == "Mrs" or x[:-1].title() == "Ms" or x[:-1].title() == "Gov" or (x[:-1].title() == "President" and new_separator.index(x) == 0):
                            mrcount+=1
                        elif x[:-1].title() == "Clinton" or x[:-1].title() == "Bush":
                            real_string = x[:-1]
                            real_p_text = ' '.join((my_text.split(x))[1:])
                            break
                        elif x[:-1].title() == "President" or x[:-1].title() == "Reagan" or x[:-1].title() == "Carter" :
                            real_string = ' '.join([x[:-1], new_separator[0]])
                            real_p_text = ' '.join((my_text.split(x))[1:])
                            break
                        

            recent_speaker = real_string
        

        if real_string:
            for name in people_array:
                    if name.split()[-1].title() in real_string.title():
                        recent_speaker = name
                        name_found = 1
                        break
            if name_found:
                content_dict[recent_speaker] = content_dict.get(recent_speaker, '') + real_p_text
            else:
                other_matches[recent_speaker] = other_matches.get(recent_speaker, '') + real_p_text
            
        else:
            if recent_speaker in content_dict.keys():
                content_dict[recent_speaker] = content_dict.get(recent_speaker, '') + my_text
            else:
                other_matches[recent_speaker] = other_matches.get(recent_speaker, '') + my_text
                    
    
    
    
    
    
    
    for k, v in content_dict.items():
        new_v = v.encode('ascii', 'ignore')
        new_v = "".join(new_v.rsplit("\""))
        new_v = "".join(new_v.rsplit("\'"))
        content_dict[k] = new_v
        
    
    dict_of_content_dicts[url_dates_dict[link]] = content_dict
    
```
